## Numpy Array And Basics

learning about Numpy :-

```python
import numpy as np
```

### Creating array from list :-

```python
arr_1d = np.array([1,2,3,4,5])
print("1D arrays",arr_1d)

arr_2d = np.array([[1,2,3],[4,5,6]])
print("2D arrays :- ")
print(arr_2d)
```

### List vs Numpy Array :-


```python
py_list = [1,2,3]
print("Python list Multiplication ", py_list*2)

np_array = np.array([1,2,3])
print("Python array Multiplication ", np_array*2)
```

### Creating Array from scratch :-

```python
zeors = np.zeros((3,4))
print("zeros array: \n",zeors,end='\n\n')

ones = np.ones((2,3))
print("Ones array: \n",ones,end='\n\n')

full = np.full((2,2),7)
print("full array: \n",full,end='\n\n')

random = np.random.random((2,3))
print("random array: \n",random,end='\n\n')

sequnece = np.arange(0,10,2)
print("Sequence array: \n",sequnece,end='\n\n')
```

### Vector , Matrix and Tensor :-


```python
vector = np.array([1, 2, 3])
print("Vector: ", vector)

matrix = np.array([[2,3,4],
                   [3,4,5]])
print("Matrix: ")
print(matrix)

tensor = np.array([[[2,3,5],[9,7,4]],
                   [[3,1,4],[4,1,4]]])
print("Tensor: ")
print(tensor)
```

### Array Properties :-



```python
arr = np.array([[1, 2, 3], 
                [4, 5, 6]])
print("Shape ",arr.shape)
print("Dimension ", arr.ndim)
print("Size ", arr.size)
print("DType ", arr.dtype)
```

### Array Reshaping :-

```python
arr = np.arange(12)
print('Original Array',arr)

reshaped = arr.reshape((3,4))
print("\nReshaped Array:- ")
print(reshaped)

flattened = reshaped.flatten()
print("\nFlatened Array :- ",flattened)

# ravel (returns view, instead of copy)
raveled = reshaped.ravel()
print("\n raveled array ", raveled)

# Transpose
transpose = reshaped.T
print("\n Transposed array ")
print(transpose)
```
